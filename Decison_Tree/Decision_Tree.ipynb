```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv")
```

```python
df
```

```python
inputs=df.drop(["salary_more_then_100k"],axis=1)
```

```python
inputs
```

```python
salary=df["salary_more_then_100k"]
```

```python
from sklearn.preprocessing import LabelEncoder
a_company=LabelEncoder()
a_job=LabelEncoder()
a_degree=LabelEncoder()
```

```python
inputs["n_company"]=a_company.fit_transform(inputs["company"])
inputs["n_job"]=a_degree.fit_transform(inputs["job"])
inputs["n_degree"]=a_degree.fit_transform(inputs["degree"])
```

```python
inputs
```

```python
inputs.drop(["company","job","degree",],axis=1,inplace=True)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs,salary)
```

```python
model.score(inputs,salary)
```

```python
model.predict([[2,2,0]])
```
