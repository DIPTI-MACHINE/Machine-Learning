```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/Exercise/carprices.csv")
df
```

```python
dummy = pd.get_dummies(df["Car Model"])
dummy
```

```python
df=pd.concat([df,dummy],axis=1)
df
```

```python
df=df.drop(["Car Model","Mercedez Benz C class"],axis=1)
```

```python
df
```

```python
x=df.drop(["Sell Price($)"],axis=1)
x
```

```python
y=df["Sell Price($)"]
```

```python
from sklearn.linear_model import LinearRegression
exe=LinearRegression()
```

```python
exe.fit(x,y)
```

```python
exe.predict([[86000,7,0,1]])
```

```python
exe.score(x,y)
```
